```python
def hanoi(n,a,b,c):
    #最简模型，移动两个盘，a所占位置表初始，b所占位置表中转，c所占位置表目的地
    #从上到下，盘序号依次为1,2,3,4,5......
    if n==1:
        print("盘1："+"{0}-->{1}".format(a,c))
    elif n==2:
        print("盘1："+"{0}-->{1}".format(a,b))#up,递归逆推到最简状况/模型
        print("盘2："+"{0}-->{1}".format(a,c))#under
        print("盘1："+"{0}-->{1}".format(b,c))#up,递归逆推到最简状况/模型
    else :
        hanoi(n-1,a,c,b)#up
        print("盘{0}：".format(str(n))+"{0}-->{1}".format(a,c))#under
        hanoi(n-1,b,a,c)#up


print("3层汉诺塔：")
hanoi(3,'a','b','c')
print("7层汉诺塔：")
hanoi(7,'a','b','c')
```

```python
def hanoi(n,origin, target, helper):
    #从上到下，盘序号依次为1,2,3,4,5......
    def move(n, fro, to, transfer):
        print(
            'from {}:{} to {}:{} with transfer {}:{}'.format(
                fro[0], fro[1:], to[0], to[1:], transfer[0], transfer[1:]
            )
        )
        to.append(fro.pop())
    
    if n==1:
        move(n, origin, target, helper)
    elif n==2:

        move(n, origin, helper, target)
        move(n, origin, target, helper)
        move(n, helper, target, origin)

    else:
        hanoi(n-1,origin,helper,target)#up
        move(n, origin, target, helper)
        hanoi(n-1,helper,target,origin)#up
    
hanoi(6, ['origin',6,5,4,3,2,1], ['target'], ['helper'])
```
