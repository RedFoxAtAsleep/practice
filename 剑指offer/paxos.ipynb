```python
from itertools import combinations, permutations
class Proposed:
    no_autoincrement = 0
    content_autoincrement = 0
    def __init__(self, kind):
        self.kind = kind
        self.no = 0
        if kind == "no":
            Proposed.no_autoincrement += 1
            self.no = Proposed.no_autoincrement
            self.val = -10+self.no
        if kind == "content":
            Proposed.content_autoincrement += 1
            self.no = Proposed.content_autoincrement
            self.val = 10-self.no
    def __repr__(self):
        return "{}{}".format(self.kind,self.no)
    
    def get_autoincrement(self):
        return Proposed.autoincrement

no1 = Proposed("no")
no2 = Proposed("no")
content1 = Proposed("content")
content2 = Proposed("content")

proposeds = [no1, no2, content1, content2]

# 完全失效的节点
# 不完全失效的节点
for i in range(1,5):
    for perm in permutations(proposeds, i):
        perm = list(perm)
        print(sum([proposed.val for proposed in perm]))
        if sum([proposed.val for proposed in perm]) <= 0:
            print(perm)
```
